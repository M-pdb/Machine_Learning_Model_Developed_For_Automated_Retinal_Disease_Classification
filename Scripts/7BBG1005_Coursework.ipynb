```python
# Install and Load Required Packages
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, BatchNormalization
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc, precision_recall_curve
from sklearn.preprocessing import label_binarize
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from sklearn.model_selection import StratifiedKFold
import os
import pandas as pd

# Define Dataset Path
data_path = "/Users/huncho/Desktop/eye-images"

# Define Image and Batch Sizes
IMG_SIZE = (224, 224)  # Aligns with CNN dimensions
BATCH_SIZE = 16
NUM_FOLDS = 5  # Number of K-Folds for Cross-Validation

# Data Augmentation - Applies transformations to improve model generalisation
datagen = ImageDataGenerator(
    rescale=1./255,  # Normalise pixel values
    rotation_range=30,  # Rotation augmentation
    height_shift_range=0.2,  # Image positioning
    brightness_range=[0.8, 1.2],  # Adjust brightness range
    shear_range=0.2,  # Minor distortions to improve generalisation
    zoom_range=0.25,  # Zoom augmentation
    horizontal_flip=True,  # Recognising mirrored images
    fill_mode='nearest'  # Filling missing pixels in transformations
)

# Load Data - Convert file paths and labels
file_paths, labels = [], []

for class_name in os.listdir(data_path):
    class_dir = os.path.join(data_path, class_name)
    if os.path.isdir(class_dir):
        for file in os.listdir(class_dir):
            file_paths.append(os.path.join(class_name, file))
            labels.append(class_name)

file_paths = np.array(file_paths)
labels = np.array(labels).astype(str)

# K-Fold Cross-Validation
kf = StratifiedKFold(n_splits=NUM_FOLDS, shuffle=True, random_state=42)
fold = 1  # Initialise fold counter

for train_index, val_index in kf.split(file_paths, labels):
    print(f"Training Fold {fold}")

    # Split dataset into training and validation sets
    train_files, val_files = file_paths[train_index], file_paths[val_index]
    train_labels, val_labels = labels[train_index], labels[val_index]

    # Ensure labels are in string format for compatibility with the generators
    train_labels = np.array(train_labels).astype(str)
    val_labels = np.array(val_labels).astype(str)

    # Use Data Augmentation for training and fixed validation generator
    train_generator = datagen.flow_from_dataframe(
        dataframe=pd.DataFrame({"filename": train_files, "class": train_labels}),
        directory=data_path,
        x_col="filename",
        y_col="class",
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=True  # Shuffle training data
    )

    val_generator = datagen.flow_from_dataframe(
        dataframe=pd.DataFrame({"filename": val_files, "class": val_labels}),
        directory=data_path,
        x_col="filename",
        y_col="class",
        target_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=False  # Ensure fixed validation order
    )

    # Define Class Weights - Assign higher weights to underrepresented classes
    class_weight_dict = {0: 1.0, 1: 2.0, 2: 3.5}

    # Load Pretrained Model (DenseNet121) - Retains pretrained features
    base_model = DenseNet121(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    base_model.trainable = True  # Allow fine-tuning

    # Freeze the first 20 layers to retain general feature representations
    for layer in base_model.layers[:20]:
        layer.trainable = False

    # Define CNN Model - Add Dense layers and Dropout for classification
    model = Sequential([
        base_model,
        BatchNormalization(),  # Stabilises Training
        GlobalAveragePooling2D(),  # Reduces feature map dimensions while retaining key information
        Dense(512, activation='relu'),  # Extract high-level features for classification
        Dropout(0.4),  # Reduces overfitting
        Dense(256, activation='relu'),
        Dropout(0.5),
        Dense(3, activation='softmax')  # Softmax outputs probability distribution over 3 classes
    ])

    # Apply a Learning Rate Scheduler and Early Stopping
    lr_callback = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-5)  # Reduce LR if validation loss stagnates
    early_stopping = EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True)  # Stops training if performance plateaus

    # Implement Focal Loss to Handle Class Imbalance
    def focal_loss(alpha=0.35, gamma=2.0):  # Alpha controls balance, gamma focuses on difficult samples
        def loss(y_true, y_pred):
            ce_loss = tf.keras.losses.CategoricalCrossentropy()(y_true, y_pred)
            p_t = tf.exp(-ce_loss)
            return alpha * (1 - p_t) ** gamma * ce_loss
        return loss

    # Compile Model
    model.compile(
        optimizer=Adam(learning_rate=0.0001),  # Prevents overshooting
        loss=focal_loss(alpha=0.35, gamma=2.0),
        metrics=['accuracy']
    )

    # Train Model
    history = model.fit(
        train_generator,
        validation_data=val_generator,
        epochs=50,
        batch_size=BATCH_SIZE,
        class_weight=class_weight_dict,
        callbacks=[lr_callback, early_stopping]  # Dynamically adjusts learning rate
    )

    # Get Predictions and True Labels
    val_labels_actual = val_generator.classes  # Integer labels
    y_pred_probs = model.predict(val_generator)  # Probability outputs
    y_pred = np.argmax(y_pred_probs, axis=1)  # Convert probabilities to class indices

    # Convert numerical indices to actual class labels
    label_map = {v: k for k, v in val_generator.class_indices.items()}
    y_pred_labels = np.array([label_map[i] for i in y_pred])  # Convert predicted indices
    val_labels_str = np.array([label_map[i] for i in val_labels_actual])  # Convert actual labels

    # Plot Training History
    def plot_training_history(history, fold):
        plt.figure(figsize=(12, 5))
        plt.subplot(1, 2, 1)
        plt.plot(history.history['accuracy'], label='Train Accuracy')
        plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
        plt.xlabel('Epochs')
        plt.ylabel('Accuracy')
        plt.legend()
        plt.title(f"Model Accuracy - Fold {fold}")

        plt.subplot(1, 2, 2)
        plt.plot(history.history['loss'], label='Train Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.legend()
        plt.title(f"Model Loss - Fold {fold}")
        plt.show()

    plot_training_history(history, fold)

    # Compute and Plot Confusion Matrix
    cm = confusion_matrix(val_labels_str, y_pred_labels, labels=list(label_map.values()))

    plt.figure(figsize=(6,6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_map.values(), yticklabels=label_map.values())
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title(f'Confusion Matrix for Fold {fold}')
    plt.show()

    # Generate Classification Report
    print(f"Classification Report for Fold {fold}")
    print(classification_report(val_labels_str, y_pred_labels, target_names=list(label_map.values())))

    # Plot ROC Curve
    y_true_binarized = label_binarize(val_labels_actual, classes=[0, 1, 2])  # Convert True Labels

    plt.figure(figsize=(8, 6))
    for i, label in enumerate(label_map.values()):
        fpr, tpr, _ = roc_curve(y_true_binarized[:, i], y_pred_probs[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f'Class {label} (AUC = {roc_auc:.2f})')

    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve for Fold {fold}')
    plt.legend()
    plt.show()

    print(f"Fold {fold} Completed Successfully")

    fold += 1  # Increment Fold Counter
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ROC-AUC values from the five folds
roc_auc_values = {
    "Fold 1": [0.93, 0.99, 0.92],
    "Fold 2": [0.88, 0.93, 0.92],
    "Fold 3": [0.95, 0.97, 0.94],
    "Fold 4": [0.83, 0.91, 0.85],
    "Fold 5": [0.95, 0.98, 0.95]
}

# Convert to NumPy array
roc_auc_array = np.array(list(roc_auc_values.values()))

# Compute mean and standard deviation
mean_roc_auc = np.mean(roc_auc_array, axis=0)
std_roc_auc = np.std(roc_auc_array, axis=0)

# Define class labels
class_labels = ["Normal", "Cataract", "Glaucoma"]

# Create the plot
plt.figure(figsize=(8, 6))

x_values = np.linspace(0, 1, 50)

# Plot mean ROC curves with confidence intervals
for i, label in enumerate(class_labels):
    plt.plot(x_values, np.full_like(x_values, mean_roc_auc[i]), 
             label=f"{label} (AUC = {mean_roc_auc[i]:.2f} ± {std_roc_auc[i]:.2f})", linewidth=2)
    plt.fill_between(x_values, mean_roc_auc[i] - std_roc_auc[i], mean_roc_auc[i] + std_roc_auc[i], 
                     alpha=0.2)

# Diagonal reference line
plt.plot([0, 1], [0, 1], linestyle="--", color="gray", alpha=0.5)

# Labels and title
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Mean ROC-AUC Curve Across Five Folds")
plt.legend()
plt.grid()

# Show the plot
plt.show()
```
